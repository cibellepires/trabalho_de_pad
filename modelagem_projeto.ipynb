# Pipeline de analise de sentimentos

Este notebook implementa o pipeline usado no trabalho de Pensamento Analítico de Dados.

Pergunta do projeto: levando em consideração o texto de uma crítica de filme, conseguimos prever automaticamente se ela expressa sentimento positivo ou negativo?

Como rodar no Google Colab:

1. envie o arquivo `IMDB Dataset.csv` para o ambiente do Colab, deixando-o em `/content/IMDB Dataset.csv`, ou coloque o arquivo em uma pasta `data`;
2. se alguma biblioteca estiver faltando, rode uma célula com `%pip install -q scikit-learn xgboost lightgbm seaborn joblib sentence-transformers`;
3. execute as células em ordem.

Como rodar localmente:

1. instale as dependências com `pip install -r requirements.txt`;
2. baixe o dataset do IMDb no Kaggle;
3. salve o arquivo como `IMDB Dataset.csv` na mesma pasta do notebook ou em uma subpasta `data`;
4. execute as células em ordem.

Fluxo do código:

1. carregar o dataset do IMDb;
2. limpar os textos das criticas;
3. separar treino e teste de forma estratificada;
4. transformar texto em números com TF-IDF;
5. treinar um baseline com Regressão Logística;
6. comparar o baseline com XGBoost, LightGBM, Random Forest e Stacking;
7. gerar embeddings semânticos com Sentence-BERT e treinar uma Regressão Logística adicional;
8. escolher o melhor modelo pelo F1-score no teste;
9. comunicar os resultados com tabela comparativa, gráficos, matriz de confusão e análise de erros.


```python
import csv
import re
from pathlib import Path

import joblib
import lightgbm as lgb
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    f1_score,
    precision_score,
    recall_score,
    roc_auc_score,
)
from sklearn.model_selection import train_test_split

RANDOM_STATE = 42
TEST_SIZE = 0.2
MODEL_PATH = Path("best_model.pkl")
VECTORIZER_PATH = Path("tfidf_vectorizer.pkl")
DATA_PATHS = [
    Path("imdb_clean.parquet"),
    Path("data/imdb_clean.parquet"),
    Path("IMDB Dataset.csv"),
    Path("data/IMDB Dataset.csv"),
    Path("/content/imdb_clean.parquet"),
    Path("/content/IMDB Dataset.csv"),
]
LABEL_MAP = {0: "Negativo", 1: "Positivo"}
TFIDF_PARAMS = {
    "max_features": 5000,
    "ngram_range": (1, 2),
    "min_df": 5,
    "max_df": 0.8,
    "stop_words": "english",
}


def read_csv_safely(data_path):
    try:
        return pd.read_csv(data_path)
    except pd.errors.ParserError as error:
        print(f"Falha na leitura padrao de {data_path}: {error}")
        print("Tentando leitura alternativa, ignorando linhas malformadas...")
        return pd.read_csv(
            data_path,
            engine="python",
            on_bad_lines="skip",
            quotechar='"',
            quoting=csv.QUOTE_MINIMAL,
        )


def load_imdb_data():
    for data_path in DATA_PATHS:
        if data_path.exists():
            if data_path.suffix == ".parquet":
                return pd.read_parquet(data_path)
            return read_csv_safely(data_path)

    if "df" in globals():
        return df.copy()

    raise FileNotFoundError(
        "Dataset nao encontrado. Baixe o IMDB Dataset do Kaggle e salve como "
        "'IMDB Dataset.csv' na pasta do notebook ou em uma subpasta 'data'."
    )


def clean_text(text):
    if not isinstance(text, str):
        return ""

    text = re.sub(r"<.*?>", " ", text)
    text = re.sub(r"[^a-zA-Z\s]", " ", text)
    return " ".join(text.lower().split())


def encode_sentiment(series):
    if pd.api.types.is_numeric_dtype(series):
        return series.astype(int)

    normalized = series.astype(str).str.strip().str.lower()
    encoded = normalized.map({"negative": 0, "positive": 1, "negativo": 0, "positivo": 1})

    if encoded.isna().any():
        unknown = sorted(normalized[encoded.isna()].unique())
        raise ValueError(f"Rotulos de sentimento nao reconhecidos: {unknown}")

    return encoded.astype(int)


df = load_imdb_data()

required_columns = {"review", "sentiment"}
missing_columns = required_columns - set(df.columns)
if missing_columns:
    raise ValueError(f"Colunas obrigatorias ausentes no dataset: {missing_columns}")

df = df.dropna(subset=["review", "sentiment"]).copy()
df["review_clean"] = df["review"].apply(clean_text)
df = df[df["review_clean"].str.len() > 0].copy()
df["sentiment_encoded"] = encode_sentiment(df["sentiment"])

print("Dados carregados")
print(f"Total de registros: {len(df)}")
print("Distribuicao das classes:")
print(df["sentiment_encoded"].map(LABEL_MAP).value_counts())

X = df["review_clean"]
y = df["sentiment_encoded"]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    stratify=y,
)

print("\nDivisao treino-teste")
print(f"Treino: {len(X_train)} registros")
print(f"Teste: {len(X_test)} registros")
print(f"Distribuicao no treino: {y_train.map(LABEL_MAP).value_counts().to_dict()}")
print(f"Distribuicao no teste: {y_test.map(LABEL_MAP).value_counts().to_dict()}")

tfidf = TfidfVectorizer(**TFIDF_PARAMS)
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

print("\nVetorizacao TF-IDF")
print(f"Matriz de treino: {X_train_tfidf.shape}")
print(f"Matriz de teste: {X_test_tfidf.shape}")
```

```python
def evaluate_model(model_name, model, X_train_model, X_test_model):
    print(f"Treinando {model_name}...")
    model.fit(X_train_model, y_train)

    y_pred = model.predict(X_test_model)
    y_proba = model.predict_proba(X_test_model)[:, 1]

    metrics = {
        "Modelo": model_name,
        "Acuracia": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1-score": f1_score(y_test, y_pred),
        "AUC-ROC": roc_auc_score(y_test, y_proba),
    }
    return model, y_pred, y_proba, metrics


models = {
    "Regressao Logistica (baseline)": LogisticRegression(
        max_iter=1000,
        random_state=RANDOM_STATE,
        n_jobs=-1,
    ),
    "XGBoost": xgb.XGBClassifier(
        n_estimators=100,
        max_depth=4,
        learning_rate=0.1,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=RANDOM_STATE,
        eval_metric="logloss",
        verbosity=0,
        n_jobs=-1,
    ),
    "LightGBM": lgb.LGBMClassifier(
        n_estimators=100,
        num_leaves=31,
        learning_rate=0.1,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=RANDOM_STATE,
        verbosity=-1,
        n_jobs=-1,
    ),
    "Random Forest": RandomForestClassifier(
        n_estimators=100,
        max_depth=10,
        min_samples_split=5,
        random_state=RANDOM_STATE,
        n_jobs=-1,
    ),
}

models["Stacking"] = StackingClassifier(
    estimators=[
        ("xgb", models["XGBoost"]),
        ("lgb", models["LightGBM"]),
        ("rf", models["Random Forest"]),
    ],
    final_estimator=LogisticRegression(max_iter=1000, random_state=RANDOM_STATE),
    cv=2,
    stack_method="predict_proba",
    n_jobs=-1,
)

trained_models = {}
predictions = {}
probabilities = {}
results = []

for model_name, model in models.items():
    trained_model, y_pred, y_proba, metrics = evaluate_model(
        model_name,
        model,
        X_train_tfidf,
        X_test_tfidf,
    )
    trained_models[model_name] = trained_model
    predictions[model_name] = y_pred
    probabilities[model_name] = y_proba
    results.append(metrics)

results_df = pd.DataFrame(results).sort_values("F1-score", ascending=False).reset_index(drop=True)
results_df
```

## Comparação adicional com embeddings

As próximas células acrescentam uma representação semântica do texto usando `sentence-transformers/all-MiniLM-L6-v2`.

Essa parte compara o TF-IDF com uma abordagem baseada em embeddings. Para manter o custo baixo no Colab, os embeddings são gerados uma vez e salvos em arquivos `.npy`. Se os arquivos já existirem, o notebook apenas carrega os vetores salvos.


```python
try:
    from sentence_transformers import SentenceTransformer
except ImportError:
    import subprocess
    import sys

    subprocess.check_call([
        sys.executable,
        "-m",
        "pip",
        "install",
        "-q",
        "sentence-transformers",
    ])
    from sentence_transformers import SentenceTransformer

import torch

EMBEDDING_MODEL_NAME = "sentence-transformers/all-MiniLM-L6-v2"
EMBEDDING_DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
EMBEDDING_BATCH_SIZE = 64 if EMBEDDING_DEVICE == "cuda" else 16
TRAIN_EMBEDDINGS_PATH = Path("X_train_embeddings_all_minilm_l6_v2.npy")
TEST_EMBEDDINGS_PATH = Path("X_test_embeddings_all_minilm_l6_v2.npy")

print(f"Modelo de embeddings: {EMBEDDING_MODEL_NAME}")
print(f"Dispositivo usado: {EMBEDDING_DEVICE}")

embedding_model = SentenceTransformer(EMBEDDING_MODEL_NAME, device=EMBEDDING_DEVICE)


def encode_or_load_embeddings(texts, output_path):
    if output_path.exists():
        embeddings = np.load(output_path)
        if embeddings.shape[0] == len(texts):
            print(f"Carregando embeddings salvos: {output_path}")
            return embeddings
        print(f"Cache ignorado porque o tamanho mudou: {output_path}")

    embeddings = embedding_model.encode(
        texts.astype(str).tolist(),
        batch_size=EMBEDDING_BATCH_SIZE,
        show_progress_bar=True,
        convert_to_numpy=True,
        normalize_embeddings=True,
    )
    np.save(output_path, embeddings)
    return embeddings


X_train_embedding_text = df.loc[X_train.index, "review"]
X_test_embedding_text = df.loc[X_test.index, "review"]

X_train_embeddings = encode_or_load_embeddings(X_train_embedding_text, TRAIN_EMBEDDINGS_PATH)
X_test_embeddings = encode_or_load_embeddings(X_test_embedding_text, TEST_EMBEDDINGS_PATH)

print(f"Embeddings de treino: {X_train_embeddings.shape}")
print(f"Embeddings de teste: {X_test_embeddings.shape}")
```

```python
embedding_classifier_name = "Embeddings + Regressao Logistica"
embedding_classifier = LogisticRegression(
    max_iter=1000,
    random_state=RANDOM_STATE,
    n_jobs=-1,
)

trained_model, y_pred, y_proba, metrics = evaluate_model(
    embedding_classifier_name,
    embedding_classifier,
    X_train_embeddings,
    X_test_embeddings,
)

trained_models[embedding_classifier_name] = trained_model
predictions[embedding_classifier_name] = y_pred
probabilities[embedding_classifier_name] = y_proba

results = [result for result in results if result["Modelo"] != embedding_classifier_name]
results.append(metrics)

results_df = pd.DataFrame(results).sort_values("F1-score", ascending=False).reset_index(drop=True)
results_df
```

```python
best_model_name = results_df.loc[0, "Modelo"]
best_model = trained_models[best_model_name]
y_pred_best = predictions[best_model_name]

print(f"Melhor modelo pelo F1-score no teste: {best_model_name}")
print("\nRelatorio de classificacao")
print(classification_report(y_test, y_pred_best, target_names=["Negativo", "Positivo"]))

cm = confusion_matrix(y_test, y_pred_best)

plt.figure(figsize=(6, 5))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=["Negativo", "Positivo"],
    yticklabels=["Negativo", "Positivo"],
)
plt.xlabel("Predito")
plt.ylabel("Real")
plt.title(f"Matriz de confusao - {best_model_name}")
plt.tight_layout()
plt.show()
```

```python
error_analysis = pd.DataFrame({
    "review_original": df.loc[X_test.index, "review"],
    "review_limpa": X_test,
    "sentimento_real": y_test.map(LABEL_MAP),
    "sentimento_predito": pd.Series(y_pred_best, index=X_test.index).map(LABEL_MAP),
})

error_analysis = error_analysis[
    error_analysis["sentimento_real"] != error_analysis["sentimento_predito"]
].copy()

print(f"Total de erros: {len(error_analysis)} de {len(X_test)} exemplos de teste")

error_analysis[["review_original", "sentimento_real", "sentimento_predito"]].head(5)
```

```python
best_model_uses_embeddings = best_model_name == "Embeddings + Regressao Logistica"

joblib.dump(best_model, MODEL_PATH)
joblib.dump(tfidf, VECTORIZER_PATH)

print(f"Modelo salvo em: {MODEL_PATH}")
print(f"Vetorizador TF-IDF salvo em: {VECTORIZER_PATH}")

if best_model_uses_embeddings:
    embedding_model_dir = Path("embedding_model_all_minilm_l6_v2")
    embedding_model.save(str(embedding_model_dir))
    print(f"Modelo de embeddings salvo em: {embedding_model_dir}")


def vectorize_new_text(text):
    if best_model_uses_embeddings:
        return embedding_model.encode(
            [str(text)],
            batch_size=1,
            convert_to_numpy=True,
            normalize_embeddings=True,
        )

    text_clean = clean_text(text)
    return tfidf.transform([text_clean])


def predict_sentiment(text, model=best_model):
    text_vector = vectorize_new_text(text)
    prediction = int(model.predict(text_vector)[0])
    probability_positive = float(model.predict_proba(text_vector)[0][1])
    confidence = probability_positive if prediction == 1 else 1 - probability_positive

    return {
        "sentiment": LABEL_MAP[prediction],
        "confidence": confidence,
    }


example_review = "This movie was absolutely amazing. I loved every minute of it."
predict_sentiment(example_review)
```

## Tabelas e gráficos para o Communicate

As próximas células organizam os resultados para a etapa de comunicação. Elas geram uma tabela interpretável, gráficos comparando F1-score e AUC-ROC, um mapa de calor das métricas e curvas ROC dos modelos.

```python
from IPython.display import display

model_info = {
    "Regressao Logistica (baseline)": {
        "Representacao": "TF-IDF",
        "Categoria": "Baseline",
    },
    "XGBoost": {
        "Representacao": "TF-IDF",
        "Categoria": "Ensemble",
    },
    "LightGBM": {
        "Representacao": "TF-IDF",
        "Categoria": "Ensemble",
    },
    "Random Forest": {
        "Representacao": "TF-IDF",
        "Categoria": "Ensemble",
    },
    "Stacking": {
        "Representacao": "TF-IDF",
        "Categoria": "Ensemble combinado",
    },
    "Embeddings + Regressao Logistica": {
        "Representacao": "Embeddings",
        "Categoria": "Modelo semantico",
    },
}

communication_results_df = results_df.copy()
communication_results_df["Representacao"] = communication_results_df["Modelo"].map(
    lambda name: model_info.get(name, {}).get("Representacao", "Nao informado")
)
communication_results_df["Categoria"] = communication_results_df["Modelo"].map(
    lambda name: model_info.get(name, {}).get("Categoria", "Nao informado")
)

communication_results_df = communication_results_df[
    ["Modelo", "Representacao", "Categoria", "Acuracia", "Precision", "Recall", "F1-score", "AUC-ROC"]
]

metric_columns = ["Acuracia", "Precision", "Recall", "F1-score", "AUC-ROC"]
communication_results_df[metric_columns] = communication_results_df[metric_columns].round(4)

display(communication_results_df)
```

```python
plot_df = communication_results_df.sort_values("F1-score", ascending=True)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.barplot(
    data=plot_df,
    y="Modelo",
    x="F1-score",
    hue="Representacao",
    dodge=False,
    ax=axes[0],
)
axes[0].set_title("Comparacao por F1-score")
axes[0].set_xlim(max(0, plot_df["F1-score"].min() - 0.03), 1.0)
axes[0].set_xlabel("F1-score")
axes[0].set_ylabel("")
axes[0].legend(title="Representacao", loc="lower right")

sns.barplot(
    data=plot_df,
    y="Modelo",
    x="AUC-ROC",
    hue="Representacao",
    dodge=False,
    ax=axes[1],
)
axes[1].set_title("Comparacao por AUC-ROC")
axes[1].set_xlim(max(0, plot_df["AUC-ROC"].min() - 0.03), 1.0)
axes[1].set_xlabel("AUC-ROC")
axes[1].set_ylabel("")
axes[1].legend(title="Representacao", loc="lower right")

plt.tight_layout()
plt.show()
```

```python
metrics_matrix = communication_results_df.set_index("Modelo")[metric_columns]

plt.figure(figsize=(10, 5))
sns.heatmap(
    metrics_matrix,
    annot=True,
    fmt=".4f",
    cmap="YlGnBu",
    cbar_kws={"label": "Valor da metrica"},
)
plt.title("Resumo das metricas por modelo")
plt.xlabel("Metrica")
plt.ylabel("Modelo")
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import roc_curve

plt.figure(figsize=(8, 6))

for model_name in communication_results_df["Modelo"]:
    y_proba = probabilities.get(model_name)
    if y_proba is None:
        continue

    fpr, tpr, _ = roc_curve(y_test, y_proba)
    auc_value = roc_auc_score(y_test, y_proba)
    plt.plot(fpr, tpr, label=f"{model_name} (AUC={auc_value:.3f})")

plt.plot([0, 1], [0, 1], linestyle="--", color="gray", label="Aleatorio")
plt.xlabel("Taxa de falsos positivos")
plt.ylabel("Taxa de verdadeiros positivos")
plt.title("Curvas ROC dos modelos avaliados")
plt.legend(loc="lower right", fontsize=8)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
baseline_name = "Regressao Logistica (baseline)"
baseline_f1 = communication_results_df.loc[
    communication_results_df["Modelo"] == baseline_name,
    "F1-score",
].iloc[0]

baseline_comparison_df = communication_results_df.copy()
baseline_comparison_df["Diferenca vs baseline"] = (
    baseline_comparison_df["F1-score"] - baseline_f1
).round(4)
baseline_comparison_df = baseline_comparison_df.sort_values(
    "Diferenca vs baseline",
    ascending=True,
)

colors = baseline_comparison_df["Diferenca vs baseline"].apply(
    lambda value: "#2ca02c" if value > 0 else "#d62728" if value < 0 else "#4c78a8"
)

plt.figure(figsize=(10, 5))
sns.barplot(
    data=baseline_comparison_df,
    y="Modelo",
    x="Diferenca vs baseline",
    palette=colors.tolist(),
)
plt.axvline(0, color="black", linewidth=1)
plt.title("Diferenca de F1-score em relacao ao baseline")
plt.xlabel("F1-score do modelo - F1-score do baseline")
plt.ylabel("")
plt.grid(axis="x", alpha=0.3)
plt.tight_layout()
plt.show()

baseline_comparison_df[
    ["Modelo", "Representacao", "Categoria", "F1-score", "Diferenca vs baseline"]
]
```
